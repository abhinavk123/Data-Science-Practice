```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
observations = 1000
```

```python
xs = np.random.uniform(low=-10,high=10,size=(observations,1))
```

```python
xz = np.random.uniform(-10,10,(observations,1))
```

```python
inputs = np.column_stack((xs,xz))
```

```python
inputs
```

```python
noise = np.random.uniform(-1,1,(observations,1))
```

```python
noise
```

```python
targets = 2*xs - 3*xz + 5 + noise
```

```python
targets
```

```python
np.savez('tf_dataset',inputs=inputs,output=targets)
```

```python
training_data = np.load('tf_dataset.npz')
```

```python
input_size = 2
output_size =1
```

```python
model = tf.keras.Sequential([
                        tf.keras.layers.Dense(output_size,
                                             kernel_initializer=tf.random_uniform_initializer(minval=-0.1, maxval=0.1),
                                                 bias_initializer=tf.random_uniform_initializer(minval=-0.1, maxval=0.1)
                                             )
])
```

```python
custom_optimizer = tf.keras.optimizers.SGD(learning_rate=0.02)
model.compile(optimizer=custom_optimizer,loss='mean_squared_error')
```

```python
model.fit(x=training_data['inputs'],y=training_data['output'],epochs=100)
```

```python
model.layers[0].get_weights()
```

```python
model.predict_on_batch(training_data['inputs']).round(1)
```
