# New Feature

```python
city_count = train['CITY_2'].value_counts()
city_count.head()
```

```python
#city_count
```

```python
def get_city_type(x):
    if x >=1000:
        return 'Metro'
    if x< 1000 and x>=500:
        return 'City'
    if x<500 and x >=100 :
        return 'Town'
    else:
        return 'Village'
    
```

```python
city_type = city_count.map(get_city_type)
```

```python
train_2 = train.copy()
test_2 = test.copy()
```

```python
for df in [train_2,test_2]:
    df['SQFT_P_ROOM']  = df['SQUARE_FT']/df['BHK_NO.']
    df['Lo+Li'] = df['LATITUDE'] + np.log(df['LONGITUDE'])
    df['Rare_House'] = df['BHK_NO.'].map(lambda x: 1 if x==5 else 0)
    df['City_Type'] = df['CITY_2'].map(lambda x: city_type[x] if x in city_type.index.values else 'Unknown')
```
