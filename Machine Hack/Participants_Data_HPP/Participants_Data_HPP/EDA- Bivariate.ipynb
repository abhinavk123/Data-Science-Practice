```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set()
```

```python
df = pd.read_csv('Univariate-Train.csv')
```

```python
df.head()
```

```python
cat_col = ['POSTED_BY', 'UNDER_CONSTRUCTION', 'RERA', 'BHK_NO.', 'BHK_OR_RK',
       , 'READY_TO_MOVE', 'RESALE']
```

```python
fig,ax = plt.subplots(3,3,figsize=(12,12))

sns.pointplot(x='POSTED_BY',y='TARGET(PRICE_IN_LACS)',data=df,ax=ax[0, 0])
sns.pointplot(x='UNDER_CONSTRUCTION',y='TARGET(PRICE_IN_LACS)',data=df,ax=ax[0, 1])
sns.pointplot(x='RERA',y='TARGET(PRICE_IN_LACS)',data=df,ax=ax[0, 2])
sns.pointplot(x='BHK_NO.',y='TARGET(PRICE_IN_LACS)',data=df,ax=ax[1, 0])
sns.pointplot(x='BHK_OR_RK',y='TARGET(PRICE_IN_LACS)',data=df,ax=ax[1, 1])
sns.pointplot(x='READY_TO_MOVE',y='TARGET(PRICE_IN_LACS)',data=df,ax=ax[1, 2])
sns.pointplot(x='RESALE',y='TARGET(PRICE_IN_LACS)',data=df,ax=ax[2, 0])
```

```python
sns.pointplot(x='BHK_NO.',y='TARGET(PRICE_IN_LACS)',hue='READY_TO_MOVE',data=df)
```
